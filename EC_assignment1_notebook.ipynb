# Barriers to Environmental Lifestyle Changes by Satisfaction Level #

## Respondents to the Opinions and Lifestyle Survey (OPN) were asked "What barriers, if any, have prevented you from making more changes to your lifestyle to help tackle environmental issues?". I am interested in how the reported barriers vary between those who expressed low satisfaction with the environment versus 'not low' satisfaction. ##

Dataset pulled from Office for National Statistics (ONS), released 14 November 2025, ONS website, statistical bulletin, [Public opinions and social trends, Great Britain: October 2025](https://www.ons.gov.uk/peoplepopulationandcommunity/wellbeing/bulletins/publicopinionsandsocialtrendsgreatbritain/october2025)

**But a more efficient and visually effective way is using graphs**

```python
import matplotlib.pyplot as plt    #import matplotlib for graphs

#various formatting for how the graphs appear
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,6)
plt.rcParams['font.size'] = 10
```

```python
low_sat = env_sat[env_sat['Environmental Satisfaction'] == 'Low']    #assigning the filtered data to a variable

#applying plot method from matplotlib to create a bar plot and telling it y and x values, colour, etc
low_sat.plot(kind='bar', x='Barriers options', y='Estimate (%)', color='#44a4a4', title='Bar plot of Barriers to Lifestyle Changes for Low Satisfaction Group')
plt.show()
```

```python
notlow_sat = env_sat[env_sat['Environmental Satisfaction'] == 'Not low']    #same as above for the other category

notlow_sat.plot(kind='bar', x='Barriers options', y='Estimate (%)', color='#0db951', title='Bar plot of Barriers to Lifestyle Changes for Not-Low Satisfaction Group')
plt.show()
```

**From these two bar plots, we can already see that the two groups have a very similar distribution. We can see this more clearly, as well as getting a clearer picture of where differences do occur, by comparing the groups side-by-side using a grouped bar plot.**

```python
x = np.arange(len(low_sat))   #create one x-axis position per barrier

plt.bar(x - 0.2, low_sat['Estimate (%)'], width=0.4, label='Low', color='#44a4a4')   #=/-2 so they're not on top of eachother
plt.bar(x + 0.2, notlow_sat['Estimate (%)'], width=0.4, label='Not low', color='#0db951')
plt.xticks(x, low_sat['Barriers options'], rotation=90)     #rotate x-axis labels due to length of text
plt.title('Comparison of Barriers by Environmental Satisfaction')
plt.legend()
```

**The grouped bar chart highlights how closely aligned the two groups are across most barrier categories. This descriptive comparison helps identify where any small differences occur and could inform further investigation in later analyses.**
